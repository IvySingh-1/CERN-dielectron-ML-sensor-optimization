```python
# Import Required Libraries

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:.3f}".format)

print("Libraries Imported Successfully")
```

```python
# Load Processed Dataset

df = pd.read_csv("../data/processed/dielectron_processed.csv")

print("Dataset Loaded Successfully")

print(df.shape)
```

```python
# Display first five rows

df.head()
```

```python
# Total Energy
# Total energy of both electrons.
# Formula:
# Total Energy = E1 + E2

df["Total_Energy"] = df["E1"] + df["E2"]

print("Feature Created : Total_Energy")
```

```python
# Energy Difference
# Absolute energy difference between the electrons.

df["Energy_Difference"] = abs(df["E1"] - df["E2"])

print("Feature Created : Energy_Difference")
```

```python
# Energy Ratio
# Ratio between electron energies.
# Small constant added to avoid division by zero.

df["Energy_Ratio"] = df["E1"] / (df["E2"] + 1e-10)

print("Feature Created : Energy_Ratio")
```

```python
# Momentum Magnitude of Electron 1
# Formula:
# sqrt(px² + py² + pz²)

df["Momentum1"] = np.sqrt(
    df["px1"]**2 +
    df["py1"]**2 +
    df["pz1"]**2
)

print("Feature Created : Momentum1")
```

```python
# Momentum Magnitude of Electron 2

df["Momentum2"] = np.sqrt(
    df["px2"]**2 +
    df["py2"]**2 +
    df["pz2"]**2
)

print("Feature Created : Momentum2")
```

```python
# Total Momentum

df["Total_Momentum"] = (
    df["Momentum1"] +
    df["Momentum2"]
)

print("Feature Created : Total_Momentum")
```

```python
# Momentum Difference

df["Momentum_Difference"] = abs(
    df["Momentum1"] -
    df["Momentum2"]
)

print("Feature Created : Momentum_Difference")
```

```python
# Total Transverse Momentum

df["Total_PT"] = df["pt1"] + df["pt2"]

print("Feature Created : Total_PT")
```

```python
# Difference in Transverse Momentum

df["PT_Difference"] = abs(
    df["pt1"] -
    df["pt2"]
)

print("Feature Created : PT_Difference")
```

```python
# PT Ratio

df["PT_Ratio"] = df["pt1"] / (df["pt2"] + 1e-10)

print("Feature Created : PT_Ratio")
```

```python
# Difference in Pseudorapidity

df["Eta_Difference"] = abs(
    df["eta1"] -
    df["eta2"]
)

print("Feature Created : Eta_Difference")
```

```python
# Difference in Phi Angle, correctly wrapped for periodicity within [-pi, pi]. Wrapping is physically required because the azimuthal angle is circular (periodic); a simple absolute difference would incorrectly yield a large value (near 2*pi) for particles close to the boundary (e.g., at +pi and -pi).

dp = df["phi1"] - df["phi2"]
df["Phi_Difference"] = np.abs(np.arctan2(np.sin(dp), np.cos(dp)))

print("Feature Created : Phi_Difference (Wrapped)")
```

```python
# Total Charge
# Since electrons usually have opposite charges,
# this feature may indicate unusual events.

df["Total_Charge"] = df["Q1"] + df["Q2"]

print("Feature Created : Total_Charge")
```

```python
# Charge Product

df["Charge_Product"] = df["Q1"] * df["Q2"]

print("Feature Created : Charge_Product")
```

```python
# Relativistic Reconstructed Mass & Delta R

# Reconstruct the invariant mass M_reco using the relativistic energy-momentum relationship. This represents the exact physical law relating energy and momentum components. We clip negative values to 0 to handle minor numerical rounding errors prior to taking the square root.
energy_sum_sq = (df["E1"] + df["E2"])**2
momentum_x_sum_sq = (df["px1"] + df["px2"])**2
momentum_y_sum_sq = (df["py1"] + df["py2"])**2
momentum_z_sum_sq = (df["pz1"] + df["pz2"])**2
mass_sq = energy_sum_sq - momentum_x_sum_sq - momentum_y_sum_sq - momentum_z_sum_sq
df["M_reco"] = np.sqrt(np.clip(mass_sq, 0, None))

# Spatial separation Delta R
df["Delta_R"] = np.sqrt(df["Eta_Difference"]**2 + df["Phi_Difference"]**2)

print("Features Created : M_reco, Delta_R")
```
