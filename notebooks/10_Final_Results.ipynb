```python
# Import required libraries
import warnings
warnings.filterwarnings('ignore')
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("Libraries loaded successfully")
```

```python
# Load model performance results
perf_path = '../outputs/model_performance.csv'
if not os.path.exists(perf_path):
    raise FileNotFoundError(f"Performance results not found at {perf_path}. Run Notebook 6 first.")
perf_df = pd.read_csv(perf_path)
print("Model performance table loaded")
```

```python
# Display best models by RMSE
best_models = perf_df.sort_values('RMSE').drop_duplicates('Model')
print("Best Performing Model Configuration per Regressor Type:")
display(best_models)
```

```python
# Plot Cost vs Accuracy Tradeoff Curve (Pareto frontier)
# We evaluate the best RMSE achieved for each feature subset size
best_subset = perf_df.groupby('Features')['RMSE'].min().reset_index()
plt.figure(figsize=(8, 5))
plt.plot(best_subset['Features'], best_subset['RMSE'], marker='o', color='purple', linewidth=2)
plt.title("Cost-Accuracy Tradeoff: Best Invariant Mass Prediction RMSE vs Sensor Feature Count")
plt.xlabel("Number of Sensors (Features) Used")
plt.ylabel("Best RMSE (GeV)")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Load and display purging metrics
purging_path = '../outputs/sensor_purging_results.csv'
if os.path.exists(purging_path):
    purging_df = pd.read_csv(purging_path)
    print("Sensor purging results loaded:")
    display(purging_df)
else:
    print("Sensor purging results not found. Run Notebook 8 first.")
```

```python
# Load and display sensor recovery metrics
pred_path = '../outputs/sensor_prediction_results.csv'
if os.path.exists(pred_path):
    pred_df = pd.read_csv(pred_path)
    print("Sensor prediction results loaded:")
    display(pred_df)
else:
    print("Sensor prediction results not found. Run Notebook 9 first.")
```
